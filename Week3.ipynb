```python
print(' Mengubah tingkat kecerahan citra')
print('---------------------------------')

try:
    brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
    print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/pengolahan_citra/fs.jpg')
brightness_image = np.zeros(original.shape, original.dtype)

#akses per piksel
for y in range(original.shape[0]):
    for x in range(original.shape[1]):
        for c in range(original.shape[2]):
            brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

#cara simple tanpa for loop
#brightness_image = cv.convertScaleAbs(original, beta=brightness)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

### **No.1**

```python
print('Mengubah kontras dan tingkat kecerahan citra')
print('--------------------------------------------')
try:
  brightness = int(input('Masukkan tingkat kecerahan [-255 - 255] : '))
  contrast = float(input('Masukkan kontras [1.0 - 3.0] : '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/pengolahan_citra/fs.jpg')
brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(contrast*original[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

 `print('Mengubah kontras dan tingkat kecerahan citra')` : mencetak teks "Mengubah kontras dan tingkat kecerahan citra" pada output.
`print('--------------------------------------------')` : mencetak teks "--------------------------------------------" pada output sebagai pemisah.
`try:` : memulai blok try untuk menangani kemungkinan terjadinya error.
`brightness = int(input('Masukkan tingkat kecerahan [-255 - 255] : '))` : meminta input berupa nilai integer untuk tingkat kecerahan dari pengguna.
`contrast = float(input('Masukkan kontras [1.0 - 3.0] : '))` : meminta input berupa nilai float untuk kontras dari pengguna.
`except ValueError:` : menangani kemungkinan error ketika pengguna memasukkan nilai yang tidak valid.
`print('Error, not a number')` : mencetak teks "Error, not a number" pada output jika terdapat error.
`original = cv.imread('/content/drive/MyDrive/pengolahan_citra/fs.jpg')` : membaca citra yang akan diolah menggunakan library OpenCV dan menyimpannya dalam variabel `original`.
`brightness_image = np.zeros(original.shape, original.dtype)` : membuat citra kosong dengan ukuran yang sama dengan citra asli menggunakan library NumPy dan menyimpannya dalam variabel `brightness_image`.
`for y in range(original.shape[0]):` : melakukan perulangan pada setiap baris citra asli menggunakan variabel `y`.
`for x in range(original.shape[1]):` : melakukan perulangan pada setiap kolom citra asli menggunakan variabel `x`.
`for c in range(original.shape[2]):` : melakukan perulangan pada setiap channel citra asli menggunakan variabel `c`.
`brightness_image[y,x,c] = np.clip(contrast*original[y,x,c] + brightness, 0, 255)` : mengubah nilai piksel pada citra kosong sesuai dengan nilai kontras dan tingkat kecerahan yang dimasukkan pengguna.
`final_frame = cv.hconcat((original, brightness_image))` : menggabungkan citra asli dan citra yang telah diolah menggunakan OpenCV dan menyimpannya dalam variabel `final_frame`.
`cv2_imshow(final_frame)` : menampilkan citra yang telah diolah pada output.

```python
print('Mengubah tingkat kecerahan citra dengan Transformasi Log')
print('--------------------------------------------')
try:
  brightness = int(input('Masukkan tingkat kecerahan : '))
except ValueError:
  print('Error, not a number')

brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(brightness*(np.log(1.0+original[y,x,c])), 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

```python
print('Lightness')
img_lgh = cv.imread('/content/drive/MyDrive/pengolahan_citra/pc.png')
img_lgh[:] = np.max(img_lgh,axis=-1,keepdims=10)/2+np.min(img_lgh,axis=-1,keepdims=10)/2
fig, ax = plt.subplots(1,2)
ax[0].imshow(img_rgb)
ax[1].imshow(img_lgh)
```

```python
print('Luminance')
img = cv.imread('/content/drive/MyDrive/pengolahan_citra/pc.png')
image = cv.cvtColor(img, cv.COLOR_BGR2RGB)
R = image[:,:,0]
G = image[:,:,1]
B = image[:,:,2]
lumi = (0.21 * R) + (0.72 * G) + (0.07 * B)

f, axarr = plt.subplots(1,2)
axarr[0].imshow(image)
axarr[1].imshow(lumi, cmap='gray')
```
