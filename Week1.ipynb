```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow 
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

urls = ["https://cdn-asset.jawapos.com/wp-content/uploads/2022/07/ferdy-sambo-komnas_1-750x593.jpg",
        "https://awsimages.detik.net.id/community/media/visual/2022/11/01/buku-hitam-ferdy-sambo-dan-map-oranye-putri-candrawathi-2_169.jpeg?w=700&q=90",
        "https://cdn1-production-images-kly.akamaized.net/gB8fyYAgHegxwC-YZQUcOaGSrLI=/1200x675/smart/filters:quality(75):strip_icc():format(jpeg)/kly-media-production/medias/4181691/original/050198800_1664958975-Ferdy_Sambo_di_Kejagung-merdeka-11.jpg"]
for url in urls:
  image = io.imread(url)                            
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)   
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)    
  final_frame = cv.hconcat((image, image_2))        
  cv2_imshow(final_frame)                           
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi * lebar = ",tinggi,"*",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255, 255, 255]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range (tinggi):
  image_4[x, int(lebar/2)] = [255, 255, 255]
final_frame = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame)
```

```python
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

thickness = 1
color = (255, 255, 255)

image_5 = cv.line(image_5, (0,0), (lebar,tinggi), color, thickness)
image_5 = cv.line(image_5, (lebar,0), (0,tinggi), color, thickness)
final_frame = cv.hconcat((image_2,image_5))
cv2_imshow(final_frame)
```
