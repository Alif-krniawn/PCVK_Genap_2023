# 2. merubah ukuran

```python
img = cv.imread('/content/drive/MyDrive/pengolahan_citra/soal.jpg')
img = cv.resize(cv.cvtColor(img,cv.COLOR_BGR2RGB),(512,1980))
plt.imshow(img)
```

# 3. membuat garis vertikal horizontal

```python
img = cv.imread('/content/drive/MyDrive/pengolahan_citra/soal.jpg')
img4 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

tinggi = img.shape[0]
lebar = img.shape[1]

for x in range (tinggi):
  img4[x, int(lebar/2)] = [255, 255, 255]
for y in range (lebar):
  img4[int((tinggi)/2),y] = [255, 255, 255]

plt.imshow(img4)
```

# 4. flip vertikal horizontal

```python
img = cv.imread('/content/drive/MyDrive/pengolahan_citra/soal.jpg')
img_t = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.imshow(img_t)
```

```python
img = cv.imread('/content/drive/MyDrive/pengolahan_citra/soal.jpg')
img_t = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),1)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.imshow(img_t)
```

# 5. menambah shape dan rectangle

```python
img = cv.imread('/content/drive/MyDrive/pengolahan_citra/soal.jpg')
cv.rectangle(img,pt1=(140,80),pt2=(190,150),color=(0,255,0),thickness=5)
cv.rectangle(img,pt1=(200,80),pt2=(250,150),color=(0,255,0),thickness=5)
cv.circle(img,center=(410,120),radius=30,color=(255,0,0),thickness=5)
cv.circle(img,center=(460,120),radius=30,color=(255,0,0),thickness=5)
cv.circle(img,center=(650,120),radius=30,color=(0,0,255),thickness=5)
cv.rectangle(img,pt1=(690,80),pt2=(720,150),color=(0,0,255),thickness=5)
plt.imshow(img)
```
